```python
# Movie Reviews
import pandas
# Read the dataset into a Dataframe called movies using Pandas
# fandango_score_comparison
movies = pandas.read_csv("fandango_score_comparison.csv", encoding = "ISO-8859-1")
# print(movies.head())
```

```python
# You can output a Dataframe as a table by typing just the variable name containing the Dataframe in the last line of a Jupyter cell. 
movies
```

```python
# Histograms
# primarily focus on the Metacritic_norm_round and the Fandango_Stars columns, 
# which will let you see how Fandango and Metacritic differ in terms of review scores.
import matplotlib.pyplot as plt
%matplotlib inline

# Create a histogram of the Fandango_Stars column.
plt.hist(movies["Fandango_Stars"])
plt.show()
```

```python
plt.hist(movies["Metacritic_norm_round"])
```

Fandango vs Metacritic Scores

There are no scores below a 3.0 in the Fandango reviews. The Fandango reviews also tend to center around 4.5 and 4.0, whereas the Metacritic reviews seem to center around 3.0 and 3.5.

```python
# Mean, Median, And Standard Deviation
# Calculate the mean of both Fandango_Stars and Metacritic_norm_round.
mean_fandango = movies["Fandango_Stars"].mean()
mean_metacritic = movies["Metacritic_norm_round"].mean()

print(mean_fandango)
print(mean_metacritic)
```

```python
# Calculate the median of both Fandango_Stars and Metacritic_norm_round.
median_fandango = movies["Fandango_Stars"].median()
median_metacritic = movies["Metacritic_norm_round"].median()
print(median_fandango)
print(median_metacritic)
```

```python
# Calculate the standard deviation of both Fandango_Stars and Metacritic_norm_round. 
import numpy
std_dev_fandango = numpy.std(movies["Fandango_Stars"])
std_dev_metacritic = numpy.std(movies["Metacritic_norm_round"])
print(std_dev_fandango)
print(std_dev_metacritic)
```

Fandango vs Metacritic Methodology

Fandango appears to inflate ratings and isn't transparent about how it calculates and aggregates ratings. Metacritic publishes each individual critic rating, and is transparent about how they aggregate them to get a final rating.

Fandango vs Metacritic number differences

The median metacritic score appears higher than the mean metacritic score because a few very low reviews "drag down" the median. The median fandango score is lower than the mean fandango score because a few very high ratings "drag up" the mean.

Fandango ratings appear clustered between 3 and 5, and have a much narrower random than Metacritic reviews, which go from 0 to 5.

Fandango ratings in general appear to be higher than metacritic ratings.
These may be due to movie studio influence on Fandango ratings, and the fact that Fandango calculates its ratings in a hidden way.

```python
# Scatter Plots
# find the largest outliers by making a scatterplot
# Make a scatterplot that compares the Fandango_Stars column to the Metacritic_norm_round column.
plt.scatter(movies["Metacritic_norm_round"], movies["Fandango_Stars"])
plt.show()
```

```python
# find the movies with the largest differences between their Metacritic and Fandango ratings.
# Subtract the Fandango_Stars column from the Metacritic_norm_round column, get the absolute value 
movies["fm_diff"] = abs(movies["Metacritic_norm_round"] - movies["Fandango_Stars"])
```

```python
# Sort movies based on the fm_diff column, in descending order.
movies.sort_values("fm_diff", inplace=True, ascending=False)
print(movies.head())
```

```python
# Correlations
# Calculate the r-value measuring the correlation between Fandango_Stars and Metacritic_norm_round 
# using the scipy.stats.pearsonr function
from scipy.stats import pearsonr
r_value, p_value = pearsonr(movies["Fandango_Stars"],movies["Metacritic_norm_round"])
print(r_value, p_value)
```

Fandango and Metacritic correlation

The low correlation between Fandango and Metacritic scores indicates that Fandango scores aren't just inflated, they are fundamentally different. For whatever reason, it appears like Fandango both inflates scores overall, and inflates scores differently depending on the movie

```python
# Use the scipy.stats.linregress function 
# create a linear regression line with Metacritic_norm_round as the x-values and Fandango_Stars as the y-values.
from scipy.stats import linregress
slope, intercept, r_value, p_value, stderr_slope = linregress(x=movies["Metacritic_norm_round"], y=movies["Fandango_Stars"])
print(slope, intercept, r_value, p_value, stderr_slope)
```

```python
# Predict what a movie that got a 3.0 in Metacritic would get on Fandango using the line.
result = slope * 3.0 + intercept
print(result)
```

```python
# Predict what a movie that got a 4.0 in Metacritic would get on Fandango using the line from the last screen.
result = slope * 4.0 + intercept
print(result)
```

```python
# Make a scatter plot using the scatter function in matplotlib.pyplot.
x = [3.0, 4.0]
y = [slope * x + intercept for x in x]
plt.scatter(x, y)


# On top of the scatter plot, use the plot function in matplotlib.pyplot to 
# plot a line using the predicted values for 3.0 and 4.0.
plt.plot(x, y)
plt.show()
```
